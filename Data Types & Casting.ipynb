```python
# In Python, the data type is set when you assign a value to a variable automaticaly. But we can also specify the data types if we want to.

# Python has the following data types built-in by default, in these categories:

#  Text Type:	str
#  Numeric Types:	int, float, complex
#  Sequence Types:	list, tuple, range
#  Mapping Type:	dict
#  Set Types:	set, frozenset
#  Boolean Type:	bool
#  Binary Types:	bytes, bytearray, memoryview """
```

```python
#Getting the data type.

# We use the type() function to find out the data type.
x = 5
y = "Five"
    
print("x = ", x)
print("y = ", y)
```

```python
# Casting

#Casting in python is therefore done using constructor functions:

# int() - constructs an integer number from an integer, a float literal (by rounding down to the previous whole number), or a string literal (providing the string represents a whole number)
# float() - constructs a float number from an integer literal, a float literal or a string literal (providing the string represents a float or an integer)
# str() - constructs a string from a wide variety of data types, including strings, integer literals and float literals 
```

```python
x = int(1)   
y = int(2.8)
z = int("3")
print(x)
print(y)
print(z)
```
